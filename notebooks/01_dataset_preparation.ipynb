```python
from datasets import load_dataset
import pandas as pd

print("Libraries imported successfully")
```

```python
dataset = load_dataset("truehealth/medicationqa")
dataset
```

```python
dataset.keys()
```

```python
dataset["train"][0]
```

```python
df = pd.DataFrame(dataset["train"])
df.head()
```

```python
df.columns
```

```python
qa_df = df[["Question", "Answer"]]

qa_df.head()
```

```python
qa_df.shape
```

```python
qa_df.isnull().sum()
```

```python
qa_df = qa_df.dropna()

qa_df.shape
```

```python
qa_df["answer_length"] = qa_df["Answer"].apply(len)

qa_df["answer_length"].describe()
```

```python
for i in range(3):
    print("QUESTION:")
    print(qa_df.iloc[i]["Question"])
    
    print("\nANSWER:")
    print(qa_df.iloc[i]["Answer"])
    
    print("\n" + "="*80 + "\n")
```

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    qa_df,
    test_size=0.2,
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42
)

print("Train:", train_df.shape)
print("Validation:", val_df.shape)
print("Test:", test_df.shape)
```

```python
train_df.to_csv("../data/train.csv", index=False)
val_df.to_csv("../data/validation.csv", index=False)
test_df.to_csv("../data/test.csv", index=False)

print("Dataset files saved successfully.")
```
